## Plot From Directory

```python
Butter_Prices={"Almond Butter":10,
              "Peanut Butter":8,
              "Cashew Butter":12}
fig, ax=plt.subplots()
ax.bar(Butter_Prices.keys(),Butter_Prices.values());
ax.set(title="Bar Plot",ylabel="Prices($)");
```

```python
book_directory = {
    "The Great Gatsby": 15.99,
    "To Kill a Mockingbird": 12.49,
    "1984": 10.99,
    "Harry Potter and the Sorcerer's Stone": 19.99,
    "The Catcher in the Rye": 14.95,
    "The Hobbit": 18.50
}
fig, ax=plt.subplots(figsize=(20,5))
fig, ax1=plt.subplots(figsize=(20,5))
fig, ax2=plt.subplots(figsize=(20,5))
fig, ax3=plt.subplots(figsize=(5,5))
ax.plot(book_directory.keys(),book_directory.values(),color="red")
ax1.scatter(book_directory.keys(),book_directory.values(),color="orange")
ax2.bar(book_directory.keys(),book_directory.values(),color="Green")
ax3.barh(list(book_directory.keys()),list(book_directory.values())
)
ax.set(title="PLOT")
ax1.set(title="SCATTER")
ax2.set(title="BAR")
ax3.set(title="HORIZONTAL BAR");
```

## Types of Subplot 

* 1st 

```python
fig, ((ax1, ax2),(ax3,ax4),(ax5,ax6))=plt.subplots(nrows=3,ncols=2,figsize=(10,12))
x=np.linspace(-10,10,100)
y=np.linspace(-np.pi,np.pi,1000)
ax1.plot(x,x**3)
ax2.plot(y,np.tan(y))
ax3.scatter(x,np.random.random(100))
ax4.bar(Butter_Prices.keys(),Butter_Prices.values())
ax5.barh(list(book_directory.keys()),list(book_directory.values()))
ax6.hist(w);
```

* 2nd

```python
fig, ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,x**3)
ax[0,1].plot(y,np.tan(y))
ax[1,0].bar(Butter_Prices.keys(),Butter_Prices.values())
ax[1,1].hist(w);
```

### Pyplot Merthod

```python
over_50=heart_diseases[heart_diseases["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter", x="age",y="chol",c="target");
```

### oo Method

```python
fig, ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax,
             cmap="winter");# this changes color
```

### Scratch oo Method

```python
fig, ax=plt.subplots(figsize=(10,6))

scatter=ax.scatter(x=heart_diseases["age"],
                   y=heart_diseases["chol"],
                   c=heart_diseases["target"],
                  cmap="winter")

ax.set(title="Cholestrol and Heart Diseases Plot",
      xlabel="Age",
      ylabel="Cholestrol")

ax.axhline(y=heart_diseases["chol"].mean(),color="red",linestyle="--")

ax.axvline(x=60,color="green",linestyle=":")

ax.legend(*scatter.legend_elements(),title="Target");
```

### Plot of age chol and thalach

```python
fig, ax=plt.subplots(figsize=(8,6))

scat=ax.scatter(x=heart_diseases["age"],
          y=heart_diseases["thalach"],
          c=heart_diseases["target"])

ax.set(title="The Thalach Target  Plot",
      xlabel="Age",
      ylabel="Thalach")

ax.axhline(y=heart_diseases["thalach"].mean(),color="red",linestyle="--")

ax.axvline(x=60,color="green",linestyle=":")

ax.legend(*scat.legend_elements(),title="target")

fig.suptitle("Heart Diseases Analysis",fontsize=16,fontweight="bold");
```
